```python
%matplotlib widget
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import ttest_ind
```

```python
table_full = pd.read_excel("2022_CarbonExperiment.xlsx",sheet_name="harvest")
```

```python
table_full
```

```python
%matplotlib widget

table = table_full.loc[table_full['fungus']=='C2']
y='dry root [mg]'
sns.boxplot(x='treatment', y='dry root [mg]', data=table)

# Calculate significance between each pair of categories
categories = table['treatment'].unique()
for i in range(len(categories)):
    for j in range(i+1, len(categories)):
        cat1 = categories[i]
        cat2 = categories[j]
        values1 = table[table['treatment'] == cat1][y]
        values2 = table[table['treatment'] == cat2][y]
        t_stat, p_value = ttest_ind(values1, values2)
        if p_value < 0.05:
            print(p_value,cat1,cat2)
            # Add an asterisk to indicate significance
            x_pos = i + (j-i)/2
            y_pos = max(values1.max(), values2.max()) + 1
            plt.text(x_pos, y_pos, '*', ha='center', va='bottom')

# Set labels and title
plt.xlabel('treatment')
plt.ylabel(y)
plt.title('Boxplot with Significance')

# Show the plot
plt.show()
```
