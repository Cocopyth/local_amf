```python
%matplotlib widget

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
r_hypha = 3 #um
density = 1.6 #mm.mm-2
density = density*100 #mm.cm-2
density = density*1e3 #um.cm-2
volume_density = density * np.pi*r_hypha**2 #um**3.cm-2 fl.cm-2
mass_density = volume_density * 1e-15 #kg.cm-2 
mass_density = mass_density * 1e9 #ug.cm-2 wet mass
percent_dry_weight = 0.1
mass_density_ir= mass_density*percent_dry_weight #ug.cm-2 dry mass
```

```python
data = pd.read_csv('Fungal_trait_data.csv')
```

```python
fig,ax = plt.subplots()
data.plot("rate.0.5",'density',kind='scatter',ax=ax)
ax.scatter(v,mass_density_ir,color="red",label="R. irregularis (A5)")
ax.scatter(v_agg,mass_density_agg,color="green",label="G. aggregatum")
ax.scatter(v_C2,mass_density_C2,color="purple",label="R. irregularis (C2)")

ax.set_xlabel('extension speed $(mm.day^{-1})$')
ax.set_ylabel('density $(\mu g.cm^{-2})$')
plt.legend()
```
