```python
import os
import shutil

# Base of the old and new paths
old_base = "C:\\Users\\coren\\AMOLF-SHIMIZU Dropbox\\DATA\\PRINCE"
new_base = "C:\\Users\\coren\\AMOLF-SHIMIZU Dropbox\\DATA\\PRINCE_ANALYSIS"

# Walk through all directories and files in the old base
for dirpath, dirnames, filenames in os.walk(old_base):
    for filename in filenames:
        # Check if the file is a .mp4 file
        if filename.endswith('.mp4'):
            # Get the full path of the file
            old_file_path = os.path.join(dirpath, filename)
            
            # Get the relative path
            relative_path = os.path.relpath(old_file_path, old_base)
            
            # Join the new base with the relative path to create the new file path
            new_file_path = os.path.join(new_base, relative_path)
            
            # Create intermediate directories in the new path
            os.makedirs(os.path.dirname(new_file_path), exist_ok=True)

            # Move the file
            shutil.move(old_file_path, new_file_path)

            print(f"Moved {old_file_path} to {new_file_path}")
            # break
```
