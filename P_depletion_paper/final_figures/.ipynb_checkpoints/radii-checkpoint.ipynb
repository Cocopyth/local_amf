```python
import CalculateBase
import CleanTubes
import Skeleton
import CalculateWidth
import CleanNodes

# Img_r06_c10 Img_r07_c14 Img_r08_c12

def Tubes(image_name, blnprint):
    print('*** START - ' + image_name + ' ***')
    # Calculates background threashold for each pixel in th eimage 
    confidencebase = CalculateBase.Generate_Base(image_name)    
    # Removes background
    CleanTubes.Clean_Tubes(image_name) 
    # skeletonize network
    Skeleton.AddSkeleton(image_name)
    # Determines sections & connetions, generates network topology & digital network image
    CalculateWidth.CalculateWidth(image_name,False)
    # Clean Nodes
    confidencenodes = CleanNodes.CleanNodes(image_name)
    # Print if Requested
    if blnprint == True:
        CalculateWidth.printtubes(image_name)
    
    return confidencebase, confidencenodes 

# Test
image_name = 'Img_r08_c12' #'Img_r04_c08'
confidencebase, confidencenodes = Tubes(image_name, True)
```

```python
import cv2

def convert_to_tiff(input_image_path, output_tiff_path):
    # Read the input image
    image = cv2.imread(input_image_path)

    # Check if the image was read successfully
    if image is None:
        print("Error: Could not read the input image.")
        return

    # Save the image as TIFF
    cv2.imwrite(output_tiff_path, image)

    print("Image saved as", output_tiff_path)

for i in [9]:
    input_image_path = "test"+str(i)+".jpg"  # Replace with the path to your input image
    output_tiff_path = "test"+str(i)+".tif"  # Specify the path for the output TIFF file

# Call the function to convert the image
convert_to_tiff(input_image_path, output_tiff_path)
```

```python
import cv2

def tif_to_png(tif_file, png_file):
    # Read the TIFF image
    image = cv2.imread(tif_file, cv2.IMREAD_UNCHANGED)
    
    # Save the image as PNG
    cv2.imwrite(png_file, image, [cv2.IMWRITE_PNG_COMPRESSION, 9])

# Example usage:
#tif_to_png("test12_Final.tif", "test12_Final.png")
for name in ['Img_r08_c12_Tubes']:#,'Img_r08_c12_Base','Img_r08_c12_Norm','Img_r08_c12_Tubes','Img_r08_c12_Thinned','Img_r08_c12_ThinnedWWidth','Img_r08_c12_ThinnedWNodes','Img_r08_c12_Superposed','Img_r08_c12_Final']:
    tif_to_png(name+'.tif', name+'.png')
```

```python
import cv2
import numpy as np

def merge_images(image1, image2):
    # Read the images
    img1 = cv2.imread(image1, cv2.IMREAD_COLOR)
    img2 = cv2.imread(image2, cv2.IMREAD_COLOR)

    # Create a black background canvas
    merged_image = np.zeros_like(img1)

    # Find non-black pixels in each image
    img1_nonblack = img1 != 0
    img2_nonblack = img2 != 0

    # Superpose non-black pixels onto the merged image
    merged_image[img1_nonblack] = img1[img1_nonblack]
    merged_image[img2_nonblack] = img2[img2_nonblack]

    return merged_image

# Paths to the TIFF images
image1_path = "blue_gradient_image.tif"
image2_path = "red_gradient_image.tif"

# Merge the images
merged_image = merge_images(image1_path, image2_path)

# Save the merged image
cv2.imwrite("merged_image.tif", merged_image)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Step 1: Load the CSV file into a DataFrame
df = pd.read_csv('outputnodes.csv', sep=';', header=None)

# Step 2: Calculate the largest value and the sum of the other two columns for each row
largest = []
murray = []

for index, row in df.iterrows():
    a, b, c = row
    l = [a-62,b-62,c-62]
    largest.append(max(l))
    l.remove(max(l))
    murray.append((l[0]**5+l[1]**5)**(1/5)+62)

# Fit a linear regression line
slope, intercept, _, _, _ = linregress(murray, largest)

# Plot the scatter plot
plt.scatter(murray, largest, color='blue')

# Plot the regression line
plt.plot(murray, slope * np.array(murray) + intercept, color='red', label='Regression Line')

plt.title('Largest value vs. Sum of other two values')
plt.xlabel('Sum of other two values')
plt.ylabel('Largest value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Load the CSV file into a Pandas DataFrame
df = pd.read_csv('data_for_figure .csv')

# Calculate the average of consecutive pairs of values in the last column
averages = []
for i in range(0, len(df)-1, 2):
    avg = (df.iloc[i, -1] + df.iloc[i+1, -1]) / 2
    averages.extend([avg, avg])

# If the number of rows is odd, add the last value to the list of averages
if len(df) % 2 != 0:
    averages.append(df.iloc[-1, -1])

# Update the last column with the calculated averages
df.iloc[:, -1] = averages

# Split the DataFrame into two separate DataFrames
first = df.iloc[:34]
second = df.iloc[34:]

# Drop duplicate rows in the second DataFrame
second = second.drop_duplicates()

first = first.drop_duplicates()
first = first.drop(6)
first.at[4,'measure'] = 10.393155

# Print the first DataFrame
print("First DataFrame:")
print(first)

# Print the second DataFrame
print("\nSecond DataFrame:")
print(second)
```
