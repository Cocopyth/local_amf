```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib widget
```

```python
data = pd.read_csv("Ineq_full_fungal_experimental_data.csv")
```

```python
data["tot_root_QD_mg"]=data["root_QD_per_mg_root_rich_origin"]+data["root_QD_per_mg_root_poor_origin"]
data["tot_root_QD"]=data["tot_root_QD_mg"]*data["total_root_bm_mg"]
data.groupby('plate').agg({"tot_root_QD":"sum","hyphal_QD_total":"sum"})
```

```python
fig,ax = plt.subplots()

select = data[data['ineq_treatment']=="Medium"]
sns.boxplot(select,x="comp",y="hyphal_bm_mg")
```
