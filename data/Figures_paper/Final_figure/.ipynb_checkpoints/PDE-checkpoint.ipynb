````python
import numpy as np    
import matplotlib.pyplot as plt

class LinearAdvection1D:
    # Matrix for LA1D 
    A=0
    # Initialization of constants 
    def __init__(self, c, x0, xN, N, deltaT,T,beta):
        self.c = c 
        self.x0 = x0   
        self.xN = xN 
        self.N = N   
        self.deltaT = deltaT   
        self.T = T     
        self.beta = beta
        # CFL number funct.   
    def CFL(self):
        deltaX= (self.xN - self.x0)/self.N
        return np.abs(self.c*self.deltaT/deltaX)
    # check CFL number <=1 or not.
    def checkCFL(self):
        if (self.CFL()<=1):
            flag=True 
        else:
            flag=False
        return flag
    # Matrix assembly of LA1D   
    def upwindMatrixAssembly_n(self):
        alpha_min=min(self.CFL(),0)
        alpha_max=max(self.CFL(),0)
        a1=[alpha_max]*(self.N-1)
        a2=[1+alpha_min-alpha_max- self.beta*self.deltaT/2]*(self.N)
        a3=[-alpha_min]*(self.N-1)
        self.A=np.diag(a1, -1)+np.diag(a2, 0)+np.diag(a3, 1)
        self.A[0,-1]=alpha_max
        self.A[N-1,0]=-alpha_min
    # Solve u=Au0
    def LaxWendroffMatrixAssembly_n(self):
        C=self.CFL();
        aleft=C*(1+C)/2.;
        amid=1.-(C*C)- self.beta*self.deltaT/2.;
        aright=C*(C-1)/2.;
        a1=[aleft]*(self.N-1)
        a2=[amid]*(self.N)
        a3=[aright]*(self.N-1)
        self.A=np.diag(a1, -1)+np.diag(a2, 0)+np.diag(a3, 1)
        self.A[0,-1]=aleft
        self.A[N-1,0]=aright  
    def Solve(self,u0):
        return np.matmul(self.A,u0) 

#############  
# Start of the code
###################

N,x0,xN,deltaT,c,T,beta=100,0.,10.,0.05,1.,0.5,0.1
# initialization of constants
LA1D = LinearAdvection1D(c, x0, xN, N, deltaT,T,beta) 

# initial value
x=np.linspace(LA1D.x0,LA1D.xN,LA1D.N)
u0=np.exp(-(x-2)*(x-2))

#plot of initial value    
plt.plot(x,u0,label="Initial value")
plt.ylabel('u')
plt.xlabel('x')
plt.legend()


# calculating solution if CFL<=1
if (LA1D.checkCFL() is True):
    print("CFL number is: ", LA1D.CFL())
    LA1D.upwindMatrixAssembly()
    for t in range(0,int(LA1D.T/LA1D.deltaT)):
        u=LA1D.Solve(u0)
        u0=u
else:
    print("CFL number is greater than 1. CFL: ", LA1D.CFL())

# ploting the last solution
plt.plot(x,u,label="Solution at t="+str(LA1D.T))
plt.legend()
plt.grid(linestyle='dotted')

plt.savefig('LA1D.png',dpi=300)

print(deltaT/((xN - x0)/N))
print(LA1D.A)
  
````

````python
%matplotlib widget

import numpy as np
import matplotlib.pyplot as plt

from matplotlib import animation, rc
from IPython.display import HTML

"""Propagating wavefront. It serves as reference and as initial condition."""

def propagatingFunction(x):
    return np.exp(-np.power(x, 2) / (2. * np.power(np.pi / 4., 2)))

"""Solution function. Rows span time, columns span space. We have ```M + 1``` rows: ```M``` iterations + ```1``` initial condition. We have ```N + 1``` columns: ```N``` discretization points + ```1``` boundary condition.
Leapfrog.
"""

def leapfrog(x, t, v1,v2, alpha,beta,dt,dx):

    rho = np.zeros(((M + 1), N + 1))
    n = np.zeros(((M + 1), N + 1))
    

    # --- Initial condition
    n[0]      = propagatingFunction(x - v1 * t[0]) 

    # --- First step
    n[1, 1 : N]  += n[0, 1 : N] #dn/dt
    n[1, 1 : N] += alpha*dt*n[0, 1 : N] #alpha n dt
    n[1, 1 : N] -= beta*dt*n[0, 1 : N]*rho[0, 1 : N] #alpha n rho dt
    n[1, 1 : N] -=  0.5 *v2* dt/dx * (n[0, 2 : N + 1] - n[0, 0 : N - 1]) #v2 dn/dx
    rho[1, 1 : N] += rho[0, 1 : N]+v2*dt*n[0, 1 : N] #v n dt
    n[1, 0]           = propagatingFunction(x[0] - v1 * t[1]) 

    Q = (1. - alpha) / (1. + alpha)
    for l in range(1, M):
        n[l+1, 1 : N]     += n[l-1, 1 : N] #dn/dt
        # ntm1 = (n[l, 1 : N]+n[l-1, 1 : N])/2
        # rhotm1 = (rho[l, 1 : N]+rho[l-1, 1 : N])/2
        ntm1 = n[l, 1 : N]
        rhotm1 = rho[l, 1 : N]
        
        n[l+1, 1 : N] += 2*alpha*dt*ntm1 #alpha n dt
        n[l+1, 1 : N] -= 2*beta*dt*ntm1*rhotm1 #alpha n rho dt
        n[l+1, 1 : N] -=  v2* dt/dx * (n[l, 2 : N + 1] - n[l, 0 : N - 1]) #v2 dn/dx *dt
        rho[l+1, 1 : N] += rho[l-1, 1 : N]+2*v2*dt*ntm1 #v n dt 
        # --- Boundary condition
        n[l + 1, 0]         = propagatingFunction(x[0] - v1 * t[l + 1])    
        n[l + 1, N]         = n[l, N - 1] - Q * n[l + 1, N - 1] + Q * n[l, N] 

    return n,rho

# """Parameters."""

# xmin  = 0                                    # --- Left boundary of the simulation domain
# xmax  = 20. * np.pi                           # --- Right boundary of the simulation domain
# t_0   = 0.                                   # --- Initial time
# t_f   = 15.                                  # --- Final time
# M     = 200                                  # --- Number of time steps
# N     = 165                                  # --- Number of space mesh points
# alpha = 10.0
# beta = 0.5
# v1 = 0.3
# v2 =  0.22
"""Space-time discretization."""

x, dx = np.linspace(xmin, xmax, N + 1, retstep = True)

t, dt = np.linspace(t_0, t_f, M + 1, retstep = True)
        

"""Define reference solution."""

X, T = np.meshgrid(x, t)           

uRef = propagatingFunction(X - v1 * T)

# """Compute solution."""

# n,rho = leapfrog(x, t, v1,v2, alpha,beta,dt,dx)

# """Animate."""


# fig, ax = plt.subplots()
# plt.close()

# ax.set_xlim(( xmin, xmax))
# ax.set_ylim((    0, 1.1))

# line1, = ax.plot([], [], lw = 2, color = "r", linestyle = 'dashed')
# line2, = ax.plot([], [], lw = 2, color = "b")

# def animate(i):
#     global x
#     global u, uRef
#     # yRef = uRef[i]
#     y    = n[i*1000]
#     # line1.set_data(x, yRef)
#     line2.set_data(x, y)
#     return (line1, line2)

# anim = animation.FuncAnimation(fig, animate, frames = (M + 1)//1000, interval = 20)
# anim.save('leapfrog.gif', fps = 30)
# rc('animation', html = 'jshtml') 
# anim
     

````

````python
def leapfrog3(r, t, alpha_1,alpha_2,dt,dx,propagatingFunction):
    mu = dt/dx

    rho = np.zeros(((M + 1), N + 1))
    n = np.zeros(((M + 1), N + 1))
    

    # --- Initial condition
    n[0]      = propagatingFunction(r -  t[0]) 
    # n[0,0]         = t[0]+1    

    # --- First step
    n[1, 1 : N]  += n[0, 1 : N] #dn/dt
    n[1, 1 : N] += (1 + dt - mu**2 + dt**2/2)*n[0, 1 : N] #alpha n dt
    # n[1, 1 : N] -= alpha2*dt*n[0, 1 : N]*rho[0, 1 : N] #alpha n rho dt
    n[1, 1 : N] -=  0.5 * dt/dx * (n[0, 2 : N + 1] - n[0, 0 : N - 1]) #v2 dn/dx
    rho[1, 1 : N] += rho[0, 1 : N]+dt*n[0, 1 : N] #v n dt
    n[1, 0]           = propagatingFunction(r[0] -  t[1]) 
    # n[1, 0]         = t[1]+1    

    Q = (1. - dt/dx) / (1. + dt/dx)
    for l in range(1, M):
        ntmj = n[l, 1:N]
        ntmjp1 = n[l, 2:N+1]
        ntmjm1 = n[l, 0:N-1]
        rj = r[1:N]
        rjm1 = r[0:N-1]
        rjp1 = r[2:N+1]
        rhotmj = rho[l, 1:N]
        rhotmjp1 = rho[l, 2:N+1]
        rhotmjm1 = rho[l, 0:N-1]
        
        n[l+1, 1:N] += ntmj
        n[l+1, 1:N] += (dt - mu**2 + dt**2/2)*ntmj #dn/dt
        # n[l+1, 1:N] += (1)*n[l-1, 1 : N] #dn/dt
        
        n[l+1, 1:N] += (-dt*mu/2 - mu/2 + mu**2/2)*ntmjp1*rjp1/rj #alpha n dt
        n[l+1, 1:N] += (dt*mu/2 + mu/2 + mu**2/2)*ntmjm1*rjm1/rj #alpha n dt
        # n[l+1, 1:N] += ( - mu/2 + mu**2/2)*ntmjp1*rjp1/rj #alpha n dt
        # n[l+1, 1:N] += ( mu/2 + mu**2/2)*ntmjm1*rjm1/rj #alpha n dt
#         n[l+1, 1 : N] -=  mu * (n[l, 2 : N + 1] - n[l, 0 : N - 1]) #v2 dn/dx *dt
        
#         n[l+1, 1 : N] -=  mu * (n[l, 2 : N + 1] - n[l, 0 : N - 1]) #v2 dn/dx *dt
        
        n[l+1, 1:N] += (- dt**2/2*alpha_2)*ntmj**2 #alpha_1 n^2 dt
        n[l+1, 1:N] += (-dt*alpha_2 - dt**2*alpha_2)*ntmj*rhotmj #alpha_2 n v dt
        n[l+1, 1:N] += (1/4*dt*mu*alpha_2)*ntmjp1*rhotmjp1*rjp1/rj #1/4 mu alpha_2 n v dt
        n[l+1, 1:N] -= (1/4*dt*mu*alpha_2)*ntmjm1*rhotmjm1*rjm1/rj #1/4 mu alpha_2 n v dt
        n[l+1, 1:N] += (1/4*dt*mu*alpha_2)*rhotmj*rjp1/rj*ntmjp1 #1/4 mu alpha_2 n+1 v dt
        n[l+1, 1:N] -= (1/4*dt*mu*alpha_2)*rhotmj*rjm1/rj*ntmjm1 #1/4 mu alpha_2 n-1 v dt
        n[l+1, 1:N] += dt**2/2*alpha_2**2*ntmj*rhotmj**2
        

        rho[l+1, 1 : N] += rhotmj +(dt - mu**2 + dt**2/2)*ntmj
        # rho[l+1, 1 : N] += rhotmj +(dt)*ntmj
        
        # --- Boundary condition
        n[l + 1, 0]         = propagatingFunction(x[0] - t[l + 1]) 
        # n[l + 1, 0]         = t[l + 1]+1    
        
        n[l + 1, N]         = n[l, N - 1] - Q * n[l + 1, N - 1] + Q * n[l, N] 

    return n,rho
````

````python
import numpy as np
from scipy.integrate import solve_ivp

# Define the parameters and initial conditions
alpha = 5
beta = 0.2
v = 0.11
L = 30.0
N = 150
x = np.linspace(0, L, N)
dx = x[1] - x[0]
n0 = 2*np.exp(-(x)**2)

# Define the function that returns the derivatives
def derivatives(t, y):
    n = y[:N]
    rho = y[N:]
    dn_dt = alpha*n - beta*n*rho - np.gradient(n*v, dx)
    drho_dt = n*v
    return np.concatenate([dn_dt, drho_dt])

# Define the time span and initial conditions for the solver
t_max = 100
t_span = [0, t_max]
y0 = np.concatenate([n0, np.zeros(N)])

# Solve the differential equation using the solve_ivp function
sol = solve_ivp(derivatives, t_span, y0, method='BDF')

# Extract the solution arrays
t = sol.t
n = sol.y[:N,:]
rho = sol.y[N:,:]

# Plot the results
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.subplot(3,1,1)
plt.plot(x, n[:,0], label='t=0')
plt.plot(x, n[:,len(t)//4], label='t=T/4')
plt.plot(x, n[:,2*len(t)//4], label='t=T/2')
plt.plot(x, n[:,3*len(t)//4], label='t=3T/4')

plt.plot(x, n[:,4*len(t)//4-1], label='t=T')
# plt.legend()
plt.ylabel('n')
plt.subplot(3,1,2)
plt.plot(x, rho[:,0], label='t=0')
plt.plot(x, rho[:,len(t)//4], label='t=T/4')
plt.plot(x, rho[:,2*len(t)//4], label='t=T/2')
plt.plot(x, rho[:,3*len(t)//4], label='t=3T/4')

plt.plot(x, rho[:,4*len(t)//4-1], label='t=T')
# plt.legend()
plt.xlabel('x')
plt.ylabel('rho')
plt.subplot(3,1,3)
plt.plot(t, np.argmax(n,axis = 0)*L/N)
pos_wave = np.argmax(n,axis = 0)
t1 = 3*len(t)//4
t2 = 2*len(t)//4
wave_velocity = (pos_wave[t1]-pos_wave[t2])/(t[t1]-t[t2])*L/N
plt.plot(t,wave_velocity*t-wave_velocity*t[t1]+np.argmax(n,axis = 0)[t1]*L/N)
# plt.legend()
plt.xlabel('t')
plt.ylabel('wavefront')
plt.show()
print(wave_velocity)
````

````python
def find_slope(alpha,beta,v,L=30,N=300,t_max = 100):
    x = np.linspace(0, L, N)
    dx = x[1] - x[0]
    n0 = np.exp(-(x)**2)

    # Define the function that returns the derivatives
    def derivatives(t, y):
        n = y[:N]
        rho = y[N:]
        dn_dt = alpha*n - beta*n*rho - np.gradient(n*v, dx)
        drho_dt = n*v
        return np.concatenate([dn_dt, drho_dt])

    # Define the time span and initial conditions for the solver
    t_max = t_max
    t_span = [0, t_max]
    y0 = np.concatenate([n0, np.zeros(N)])

    # Solve the differential equation using the solve_ivp function
    sol = solve_ivp(derivatives, t_span, y0, method='BDF')

    # Extract the solution arrays
    t = sol.t
    n = sol.y[:N,:]
    rho = sol.y[N:,:]
    pos_wave = np.argmax(n,axis = 0)
    t1 = 3*len(t)//4
    t2 = 2*len(t)//4
    wave_velocity = (pos_wave[t1]-pos_wave[t2])/(t[t1]-t[t2])*L/N
    return(wave_velocity)
````
