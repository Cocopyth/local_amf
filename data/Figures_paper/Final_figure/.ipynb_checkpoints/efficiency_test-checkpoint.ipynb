```python
import networkx as nx
import matplotlib.pyplot as plt
G = nx.Graph()
G.add_nodes_from([1, 2, 3, 4])
G.add_weighted_edges_from([(1,2,1),(3,1,1),(2, 4,1)])
G.add_nodes_from([5,6])
G.add_weighted_edges_from([(3,4,1),(5,3,1),(6, 4,1)])
```

```python
nx.draw(G, with_labels=True, node_color='lightblue', node_size=800, font_weight='bold')
plt.show()
```

```python
def make_graph(n):
    if n == 0:
        G = nx.Graph()
        G.add_nodes_from([1, 2, 3, 4])
        G.add_weighted_edges_from([(1,2,1),(3,1,1),(2, 4,1)])
    else:
        G,pos = make_graph(n-1)
        G.add_nodes_from([3+2*n,4+2*n])
        G.add_weighted_edges_from([(3+2*(n-1),4+2*(n-1),1),(3+2*(n-1),3+2*(n),1),(4+2*(n-1),4+2*(n),1)])
    pos =  {node : ((node-1)//2,node%2) for node in G}
    return(G,pos)
```

```python
def make_delaunay_graph(n):
    G,pos = make_graph(n)
    G.add_weighted_edges_from([(3+2*n,4+2*n,1)])
    G.add_weighted_edges_from([(3+2*k,4+2*(k-1),1) for k in range(n+1)])

    return(G,pos)
```

```python
G,pos = make_graph(4)
nx.draw(G,pos, with_labels=True, node_color='lightblue', node_size=800, font_weight='bold')
plt.show()
```

```python
G,pos = make_delaunay_graph(4)
nx.draw(G,pos, with_labels=True, node_color='lightblue', node_size=800, font_weight='bold')
plt.show()
```

```python
T = nx.minimum_spanning_tree(G)
nx.draw(T,pos, with_labels=True, node_color='lightblue', node_size=800, font_weight='bold')
plt.show()
```
