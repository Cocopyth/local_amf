```python
from shapely.geometry import LineString, MultiPolygon, Polygon
from shapely.ops import split

def splitPolygon(polygon, dx, dy):
    minx, miny, maxx, maxy = polygon.bounds
    nx = int((maxx - minx) / dx)+1
    ny = int((maxy - miny) / dy)+1
    horizontal_splitters = [LineString([(minx, miny + i*dy), (maxx, miny + i*dy)]) for i in range(ny)]
    vertical_splitters = [LineString([(minx + i*dx, miny), (minx + i*dx, maxy)]) for i in range(nx)]
    splitters = horizontal_splitters + vertical_splitters
    result = polygon
    
    for splitter in splitters:
        result = MultiPolygon(split(result, splitter))
    
    return result
```

```python
geoms = splitPolygon(shape0,100,100).geoms
```

```python
skeletons = []
for skeleton in exp.skeletons:
    if skeleton is None:
        skeletons.append({})
    else:
        skeletons.append(skeleton)
exp.multipoints = [gpd.GeoSeries([Point(pixel) for pixel in skeleton.keys()]) for skeleton in skeletons]  
```

```python
def get_length_shape(exp,shape,t):
    skeleton = exp.skeletons[t]
    intersect = exp.multipoints[t].loc[exp.multipoints[t].within(shape)]
    tot_length = len(intersect)*1.725
    return(tot_length)
```

```python
get_length_shape(exp,shape0,t)
```

```python
np.sum([get_length_shape(exp,geom,t) for geom in geoms])
```

```python
densities = [get_length_shape(exp,geom,t)/geom.area for geom in geoms if abs(geom.area-max_area)<1e-2]
```

```python
import scipy
n_resamples = 10
res = scipy.stats.bootstrap((np.array(densities),),np.mean,
                        vectorized = True,
                        method="basic",
                        n_resamples = n_resamples)
```
