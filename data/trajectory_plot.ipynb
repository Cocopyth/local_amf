```python
def get_filename(hyph,rand):
    if rand:
        return(f'branches/ts_{hyph}_auto_rand.npy')
    else:
        return(f'branches/ts_{hyph}_auto.npy')
```

```python
def time_since_last_branch(row,rand=False):
    t = row['time_since_emergence']
    hyph = row['end']
    table = to_plot.loc[to_plot['end']==hyph]
    t_real = np.load(get_filename(hyph,rand))
    t_real.sort()
    if len(t_real)==0:
        return(None)
    if t<t_real[0]:
        return(None)
    for index,tb in enumerate(t_real):
        if t<tb:
            return(t-t_real[index-1])
    return(t-t_real[-1])
        
```

```python
def normalised_speed_since_last_branch(row,rand=False,y='speed'):
    t = row['time_since_emergence']
    hyph = row['end']
    table = to_plot.loc[to_plot['end']==hyph]
    t_real = np.load(get_filename(hyph,rand))
    t_real.sort()
    if len(t_real)==0:
        return(None)
    if t<t_real[0]:
        return(None)
    for index,tb in enumerate(t_real):
        if t<tb:
            if len(table.loc[table['time_since_emergence']==t_real[index-1]])>0:
                speed_branch = np.float(table.loc[table['time_since_emergence']==t_real[index-1]][y])
                if speed_branch>100:
                    return(row[y]/speed_branch)
                else:
                    return(None)
    if len(table.loc[table['time_since_emergence']==t_real[-1]])>0:
        speed_branch = np.float(table.loc[table['time_since_emergence']==t_real[-1]][y])
        if speed_branch>100:
            return(row[y]/speed_branch)
        else:
            return(None)
```

```python
def ZMUV_speed_since_last_branch(row,rand=False,y='speed_residual_ZMUV'):
    t = row['time_since_emergence']
    hyph = row['end']
    table = to_plot.loc[to_plot['end']==hyph]
    t_real = np.load(get_filename(hyph,rand))
    t_real.sort()
    if len(t_real)==0:
        return(None)
    if t<t_real[0]:
        return(None)
    for index,tb in enumerate(t_real):
        if t<tb:
            if len(table.loc[table['time_since_emergence']==t_real[index-1]])>0:
                return(row[y])
    if len(table.loc[table['time_since_emergence']==t_real[-1]])>0:
        return(row[y])
```

```python
def normalised_speed_until_next_branch(row,rand=False,y='speed'):
    t = row['time_since_emergence']
    hyph = row['end']
    table = to_plot.loc[to_plot['end']==hyph]

    t_real = np.load(get_filename(hyph,rand))
    t_real.sort()    
    if len(t_real)==0:
        return(None)
    if t<=t_real[0]:
        if len(table.loc[table['time_since_emergence']==t_real[0]])>0:
            speed_branch = np.float(table.loc[table['time_since_emergence']==t_real[0]][y])
            if speed_branch>100:
                return(row[y]/speed_branch)
            else:
                return(None)
    for index,tb in enumerate(t_real):
        if t<=tb:
            if len(table.loc[table['time_since_emergence']==tb])>0:
                speed_branch = np.float(table.loc[table['time_since_emergence']==tb][y])
                if speed_branch>100:
                    return(row[y]/speed_branch)
                else:
                    return(None)
    return(None)
```

```python
def ZMUV_speed_until_next_branch(row,rand=False,y='speed_residual_ZMUV'):
    t = row['time_since_emergence']
    hyph = row['end']
    table = to_plot.loc[to_plot['end']==hyph]
    t_real = np.load(get_filename(hyph,rand))
    t_real.sort()    
    if len(t_real)==0:
        return(None)
    if t<=t_real[0]:
        if len(table.loc[table['time_since_emergence']==t_real[0]])>0:
            return(row[y])
    for index,tb in enumerate(t_real):
        if t<=tb:
            if len(table.loc[table['time_since_emergence']==tb])>0:
                return(row[y])
    return(None)
```

```python
def time_until_next_branch(row,rand=False):
    t = row['time_since_emergence']
    hyph = row['end']
    table = to_plot.loc[to_plot['end']==hyph]
    t_real = np.load(get_filename(hyph,rand))
    t_real.sort()    
    if len(t_real)==0:
        return(None)
    if t<t_real[0]:
        return(t_real[0]-t)
    for index,tb in enumerate(t_real):
        if t ==tb:
            return(0)
        if t<tb:
            return(tb-t)
    return(None)
```
