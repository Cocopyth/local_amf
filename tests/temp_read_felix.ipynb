```python
%matplotlib widget

import pandas as pd
table = pd.read_csv('WAVE0002.csv',skiprows = 14,delimiter=",")
```

```python
table
```

```python
labels = ['left next to pos 3','right between pos 38 and 39','left between pos 5 and 6', 'right next to pos 37','bottom right','bottom left']
```

```python
rolling_mean = 1000
for i in range(1,7):
    table[labels[i-1]] =table[f"CH-{i}[C]"].rolling(rolling_mean).mean()
```

```python
table['Time_h']=table['Time']/3600
```

```python
table.plot("Time_h",[labels[i-1] for i in range(1,7)],alpha= 0.2,ylim=(24,26))
```
